## Machine Learning Systems Classification

### Reinforcement Learning
The agent observes the environment, select and perform actions, and get rewards in return (or penalties in the form of negative rewards). It must then learn by itself what is the best strategy, called a policy, to get the most reward over time.


### Supervised Learning
In supervised training, the training data you feed to the algorithm includes desired solutions, called <i>labels</i>. A typical supervised learning task is <b>classification</b>
### Unsupervised Learning
In unsupervised learning, the training data is unlabeled

## Reinforcement Learning

RL can be applied to a wide variety of tasks. Here are a few examples
<ol>
1. The agent can be the program controlling a walking robot. The environment being the real world which the agent observes through a set of sensors such as cameras and touch sensors. It may be programmed to get <b>positive rewards</b> whenever it <b>approaches the target destination</b> and <b> negative rewards </b> whenever <b>it wastes time, goes in the wrong direction, or falls down</b>
</ol>

<ol>
2. The agent can be the program controlling Ms. Pac-Man. In this case the environment is a simulation of the Atari game, the actions are the nine possible joystick positions, the observations are screenshots, and the rewads are just game points</ol>

<ol>
3. The agent does not have to control a physically (or virtually) moving thing. It can be a smart thermostat, getting rewards when humans need to tweak the temperature, so the agent must learn to anticipate human needs</ol>

<ol>
4. The agent can observe stock market prices and decide how much to buy or sell every second. Rewards are obviously monetargy gains and losses.</ol>

Note that there may not be any positive rewards at all.

## Policy Search

The algorithm used by the software agent to determine its actions is called <i>policy</i>


When the policy space is too large (which is generally the case), one may think of other ways than brute forcing it. Another way to explore the policy space is to use <i> genetic algorithms </i>. For example, you could randomly create a first generation of 100 policies and try them out, then "kill" the 80 worst policies and make the 20 survivors produce 4 offspring each. An offspring is just a copy of its parent PLUS some random variation. You can continue to iterate through generations this way, until you find a good policy
